## Exercise 1

```python
#initialize grids
xgrid = np.linspace(0, 6, 100)
ygrid = np.linspace(0, 6, 100)
X, Y = np.meshgrid(xgrid, ygrid)

#Get constraints in form y(x), x
const1 = lambda x: (-4 - 2*x) / -3
const2 = lambda x: (1 - x)/-6
const3 = lambda x: 6 - x

#Define func
f = lambda x, y: 5*x-4*y

const1grid= np.array(const1(xgrid))
const2grid= np.array(const2(xgrid))
const3grid= np.array(const3(xgrid))

plt.plot(xgrid, const1grid, label = "2x - 3y = -4")
plt.plot(xgrid, const2grid, label = "x - 6y = 1")
plt.plot(xgrid, const3grid, label = "x + y = 6")
plt.plot(xgrid, np.zeros(100), label = "x = 0")
plt.plot(np.zeros(100), ygrid, label = "y = 0")
plt.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))
plt.contourf(X, Y, f(X, Y), cmap ="viridis")
plt.plot(6 - 5/7, 5/7, 'o', color = "blue")
plt.text(2, 1, "P", fontsize =20, color = "black")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Constrained Optimum of 5x-4y')
plt.colorbar()
```

The region labeled "P" is the feasible set, the function values are colormapped, and the optimal point is $(6-\frac{5}{7},\frac{5}{7} )$

## Problem 2

Everywhere I have $x_2 = y$.

```python
#initialize grids
xgrid = np.linspace(-1, 10, 100)
ygrid = np.linspace(-5, 7, 100)
X, Y = np.meshgrid(xgrid, ygrid)

#Get constraints in form y(x), x
const1 = lambda x: (15 - x) / 3
const2 = lambda x: (18 - 2*x)/3
const3 = lambda x: -4 + x

#Define func
f = lambda x, y: 3*x+y

const1grid= np.array(const1(xgrid))
const2grid= np.array(const2(xgrid))
const3grid= np.array(const3(xgrid))

plt.plot(xgrid, const1grid, label = "x + 3y = 15")
plt.plot(xgrid, const2grid, label = "2x + 3y = 18")
plt.plot(xgrid, const3grid, label = "x - y = 4")
plt.plot(xgrid, np.zeros(100), label = "x = 0")
plt.plot(np.zeros(100), ygrid, label = "y = 0")
plt.legend(loc='center left', bbox_to_anchor=(1.2, 0.5))
plt.contourf(X, Y, f(X, Y), cmap ="viridis")
plt.plot(6, 2, 'o', color = "blue")
plt.text(2, 1, "P", fontsize =20, color = "black")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Constrained Optimum of 3x + y')
plt.colorbar()
```

P is the feasible set. The vertices are $(0, 0), (0, 4), (6, 2), (3, 4),$ and $(0, 5)$. The optimum is attained at the blue point $(6, 2)$.
