```python
import numpy as np
from matplotlib import pyplot as plt
import pydataset as data
```

### birthwt

```python
df = data.data('birthwt')

dfrace = df.groupby('race')
dfrace['bwt'].mean().plot(kind = 'barh')
plt.xlabel("Average Birth Weight")
plt.ylabel("Race")
plt.xlim(0, 3500)
plt.show()

dfrace = df.groupby('race')
dfrace['age'].mean().plot(kind = 'barh')
plt.xlabel("Average age at time of giving birth")
plt.ylabel("Race")
plt.xlim(0, 35)
plt.show()

df.plot('age', 'bwt', kind = 'scatter')
plt.xlabel("Age")
plt.ylabel("Birth Weight")
plt.show()
```

### road

```python
df = data.data('road')
df = df.sort_values('deaths')
df.plot(y= 'deaths', kind= 'barh')
plt.title("Traffic Deaths by State")
plt.xlabel("Traffic Deaths")
plt.ylabel("State")

df.plot(y = 'popden', kind = 'hist')
plt.title("Population Density Frequency (including DC)")
plt.xlabel("Population Density (people per square mile)")
plt.ylabel('Frequency')
df=df.drop('DC')
df.plot(y = 'popden', kind = 'hist')
plt.title("Population Density Frequency (Excluding DC)")
plt.xlabel("Population Density (people per square mile)")
plt.ylabel('Frequency')
```

The bar chart simply shows how many traffic deaths occured in each state - there is much variation. The histograms suggest that including DC in our dataset will throw any analyses involving population density haywire.

### bfeed

```python
df = data.data('bfeed')
df['duration'].plot(kind = 'hist', title="Histogram of length of breastfeeding")
plt.xlabel("weeks of breastfeeding")
plt.show()

df.groupby('poverty').mean()['duration'].plot(kind = 'barh', title = 'Average length of breastfeeding by poverty (1 = poverty)')
plt.xlabel('Weeks of breastfeeding')
plt.show()

df.groupby('ybirth').mean()['duration'].plot(kind = 'barh', title = 'Average length of breastfeeding by Year')
plt.xlabel('Weeks of breastfeeding')
plt.show()
```

### lung

```python
df = data.data('lung')
df.groupby('status').mean()['age'].plot(kind = 'barh', title = 'Average age of person with a given status')
plt.ylabel('Satus')
plt.xlabel('Average age of people with this status')
plt.show()

df.plot('meal.cal', 'wt.loss', kind = 'scatter')
plt.xlabel('Meal Calories')
plt.ylabel('Weight Loss')
plt.show()

df[(df['status'] == 2)].groupby('sex').count()['inst'].plot(kind = 'barh', title= 'Number of people of each sex with status = 2')
plt.ylabel('sex')
plt.xlabel('number of instances of status 2')
```

### heart

```python
df = data.data('heart')

df.groupby('id').mean().plot('year', 'age', kind = 'scatter', title = "Year vs. Age (for a given ID)")
plt.show()

df.groupby('event').mean()['transplant'].plot(kind= 'barh', title = 'Probability of getting a transplant at event 1 or 0')
plt.show()

df['stop'].plot(kind = 'hist', title = 'Histogram of stop values')
plt.show
```
