```python
import numpy as np
from matplotlib import pyplot as plt

def ex1func(n):
    '''
    This function accepts an integer n, creates an nxn matrix of samples
    from the normal distribution with mean 0, st. dev. 1, computes the
    mean of each row of the matrix, and computes the variance of these
    means.
    '''
    matrix = np.random.normal(size=(n, n))
    means = np.sum(matrix, axis = 1) / n
    var = np.var(means)
    return var

def ex1func2():
    '''
    This function creates an array of the integers, 100, 200, ..., 1000
    and the corresponding outputs of ex1func().
    '''
    array = np.zeros((2, 10))
    array[0] = np.linspace(100, 1000, num = 10, dtype=int)
    for i in range(0, len(array[0])):
        array[1, i] = ex1func(int(array[0, i]))
    return array

# Show Plot
data = ex1func2()
plt.plot(data[0], data[1])
plt.show()
```

## Problem 2

```python
#fineness parameter
d = 200

def ex2func():
    ''' 
    This function has no arguments, but it creates a plot of the 
    functions sin(x), cos(x), and arctan(x) on the domanin [-2pi, 2pi].
    A "fineness parameter" is defined above
    '''
    #initialize grid
    X = np.linspace(-2 * np.pi, 2 * np.pi, d)
    #Calculate function output arrays
    sinX = np.sin(X)
    cosX = np.cos(X)
    arctanX = np.arctan(X)
    #Stack into one array
    Data = np.vstack((X, sinX, cosX, arctanX))
    # Plot
    plt.plot(Data[0], Data[1])
    plt.plot(Data[0], Data[2])
    plt.plot(Data[0], Data[3])
    plt.show()
ex2func()
```
